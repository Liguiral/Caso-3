```python
del df["Taxes"] #borra la columna de taxes porque no tenía muchos datos registrados
```

```python
df
```

```python
df.dropna(thresh=2) #borra los rows con menos de dos datos
```

```python
df=df.replace("--",np.nan) #para reemplazar las celdas con valores "--" por "NaN"
df
```

```python
df.Gender=df.Gender.replace(np.nan,"U") #Reemplazar los datos Nulos por "U" en la columna de genero
df.Gender.unique()
```

```python
df.Department=df.Department.replace(np.nan,"Bench") #Reemplazar los datos Nulos por "Bench" en la columna de Department
df.Department.unique()
```

```python
df.HourlyRate=pd.to_numeric(df.HourlyRate) #Reemplazar los valores nulos de la columna “HourlyRate” por la mediana
df.HourlyRate=df.HourlyRate.replace(np.nan,df.HourlyRate.median())
df.isnull().sum()
```

```python
df.OverTime=df.OverTime.replace(["0","1"],np.nan)
df.OverTime=df.OverTime.replace(np.nan,"No")
df.OverTime.unique() #Se reemplazaron 0 y 1, por nan y luego nan por "No"
```

```python
df.to_csv("Processeddata.csv",index=False)
```
